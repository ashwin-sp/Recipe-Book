```python
zero_cols = [ col for col, is_zero in ((recipes == 0).sum() == recipes.shape[0]).items() if is_zero ]
recipes.drop(zero_cols, axis=1, inplace=True)
recipes
```

```python
#dessert_recipes = recipes.iloc[: , :7]
dessert_recipes = recipes.loc[: ,['rating','calories','protein','fat','sodium','#cakeweek','#wasteless','dessert']]
str(dessert_recipes)
dessert_recipes.describe()
```

```python
dessert_recipes = dessert_recipes[dessert_recipes['calories'] <= 20000]
dessert_recipes = dessert_recipes[dessert_recipes['sodium'] <= 20000]
dessert_recipes.describe()
```

```python
dessert_recipes.fillna(dessert_recipes.mean(), inplace=True)
dessert_recipes.describe()
```

```python
dessert_recipes.isnull().values.any()
dessert_recipes
```

```python
ggplot(dessert_recipes, aes(x='calories', y='fat')) + geom_point()
```
